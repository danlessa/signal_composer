#  Walking through `signal_composer`

```python
import sys
sys.path.append("../")
```

```python
from signal_composer import Signal
import matplotlib.pyplot as plt
import numpy as np
```

```python
def plot_signal(signal, samples=100):
    x = list(range(samples))
    y = signal(samples)
    plt.plot(x, y)
    plt.show()
```

## Simple Signals

### From Arrays

```python
signal = Signal([1, 0, 1, -1, 1])
plot_signal(signal)
```

### From Functions

```python
f = lambda x: x ** 3 - x ** 2
signal = Signal(f)
plot_signal(signal)
```

```python
f = lambda x: x ** 3 - x ** 2
signal = Signal(f, function_input_span=[0, 1 + 0.25])
plot_signal(signal)
```

## Composing Signals

### From adding together

```python
signal_1 = Signal([1, 1])
signal_2 = Signal([1, 0])
composed_signal = signal_1 + signal_2
plot_signal(composed_signal)
```

```python
signal_1 = Signal([0, 1])
signal_2 = Signal(lambda x: 1 + x ** 3)
composed_signal = signal_1 + signal_2

N_points = 100
x = list(range(N_points))
y = composed_signal(N_points)
plt.plot(x, y)
plt.show()
```

```python
signal_1 = Signal([0, 1])
signal_2 = Signal(lambda x: x ** (1/3), function_input_span=[1, 100])
composed_signal = signal_1 + signal_2
plot_signal(composed_signal)
```

### From lists

```python
signals = [
    (1, 1),
    (1, 0),
    (0, 1)
]

composed_signal = Signal.compose_from_list(signals)
plot_signal(composed_signal)
```

```python
signals = [
    lambda x: x ** 2,
    [-1, 0, 1],
    lambda x: 4 + x ** (1/2)
]

composed_signal = Signal.compose_from_list(signals)
plot_signal(composed_signal)
```

```python
signals = [
    lambda x: 1 + 5 * np.random.random() * x,
    [1, -2, 1],
    lambda x: 4 + x ** (1/2)
]

composed_signal = Signal.compose_from_list(signals)
plot_signal(composed_signal)
```
